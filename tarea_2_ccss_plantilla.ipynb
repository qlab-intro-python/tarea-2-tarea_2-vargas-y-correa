8. Calcula los retrasos promedio por destino para vuelos que se originan en Nueva York (NYC). Crea una variable que clasifique los destinos de peor a mejor, donde 1 representa el peor lugar en el ranking.







```python
# Solución
flights_nyc = flights.loc[flights.origin == 'JFK']
print(flights_nyc)
```

```python
#Primero quitar los null
null_sum = flights_nyc.isna().sum()
flights_nyc.columns[ null_sum < len( flights_nyc ) * 0.2 ] 
flights_nyc.drop(columns = flights_nyc.columns[null_sum > len( flights_nyc ) * 0.2 ], inplace = True )
```

```python
#Ahora si, calcular promedio por destino
mean_arr_delay=flights_nyc.groupby("dest")["arr_delay"].mean()
mean_arr_delay
```

```python
#Creando una clasificación de 1 a 5, donde 1 es el peor y 5 el mejor (utilizando pd.qcut para ordenar según quintiles)
mean_arr_delay_clasificado = pd.qcut(mean_arr_delay, 5, labels=[1, 2, 3, 4, 5])
print(mean_arr_delay_clasificado)
```
